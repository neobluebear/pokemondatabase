```python
def fetch_pokemon_data(pokemon):
    """Fetches data from PokeAPI for a given Pokemon."""
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon.lower()}/"
    response = requests.get(url)
    return response.json()
```

```python
def insert_pokemon(pokemon_id, pokemon_data):
    """Processes and inserts Pokemon types into the database."""
    types = []
    for data in pokemon_data['types']:
        type_name = data['type']['name']
        type_id = get_type_id(type_name)  
        types.append(type_id)
    
    if len(types) == 1:
        sql = "INSERT INTO pokemon(pokemon_id, name, type1_id) VALUES (%s, %s, %s)"
        cursor.execute(sql, (pokemon_id, pokemon_data['name'], types[0]))
    elif len(types) == 2:
        sql = "INSERT INTO pokemon(pokemon_id, name, type1_id, type2_id) VALUES (%s, %s, %s, %s)"
        cursor.execute(sql, (pokemon_id, pokemon_data['name'], types[0], types[1]))
```

```python
def insert_pokemon_stats(pokemon_id, pokemon_data):
    """Processes and inserts Pokemon stats into the database."""
    stats = {data['stat']['name']: data['base_stat'] for data in pokemon_data['stats']}
    sql_stats = "INSERT INTO pokemon_stats(pokemon_id, hp, attack, defense, special_attack, special_defense, speed) VALUES (%s, %s, %s, %s, %s, %s, %s)"
    cursor.execute(sql_stats, (pokemon_id, stats['hp'], stats['attack'], stats['defense'], stats['special-attack'], stats['special-defense'], stats['speed']))
```

```python
def insert_pokemon_abilities(pokemon_id, pokemon_data):
    """Processes and inserts Pokemon abilities into the database."""
    for data in pokemon_data['abilities']:
        name = ' '.join(word.capitalize() for word in data['ability']['name'].split('-'))
        ability_url = data['ability']['url']
        ability_data = requests.get(ability_url).json()
        #description = ability_data['flavor_text_entries'][0]['flavor_text']
        
        english_description = next((entry for entry in ability_data["effect_entries"] if entry["language"]["name"] == "en"), None)
        if english_description is not None:
            description = english_description['short_effect']
        else:
            description = "Description not found"   
        
        sql_find_ability = "SELECT ability_id FROM abilities WHERE name = %s AND description = %s"
        cursor.execute(sql_find_ability, (name, description))
        ability_id = cursor.fetchone()
        
        if ability_id:
            sql_map_ability = "INSERT INTO pokemon_abilities(pokemon_id, ability_id) VALUES (%s, %s)"
            cursor.execute(sql_map_ability, (pokemon_id, ability_id[0]))
        else:
            sql_insert_ability = "INSERT INTO abilities(name, description) VALUES (%s, %s)"
            cursor.execute(sql_insert_ability, (name, description))
            cursor.execute(sql_find_ability, (name, description))
            ability_id = cursor.fetchone()
            sql_map_ability = "INSERT INTO pokemon_abilities(pokemon_id, ability_id) VALUES (%s, %s)"
            cursor.execute(sql_map_ability, (pokemon_id, ability_id[0]))
```

```python
def populate_db(pokemons):
    """Main function to process a list of Pokemon names."""
    for pokemon in pokemons:
        pokemon_data = fetch_pokemon_data(pokemon)
        pokemon_id = pokemon_data['id']
        print(pokemon_id, pokemon_data['name'])
        
        insert_pokemon(pokemon_id, pokemon_data)
        insert_pokemon_stats(pokemon_id, pokemon_data)
        insert_pokemon_abilities(pokemon_id, pokemon_data)
        pokemon_db.commit()
    
```
